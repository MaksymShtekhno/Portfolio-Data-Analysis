```
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Plot configurations
```

```
df = pd.read_csv('movies.csv')
```

```
# 1) Find missing data

#With a NupPy library it is possible to find a relative amount of data, that is missing in every column
#To find it lets iterate through elery column, and apply a isnull() method to a variable to see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```
# Now we want to create a new column with a year of release of the film, because the old one ist not that correct

df['yearcorrect'] = df['released'].astype(str).str[:4] #Create a new column 'yearcorrect' that contains first 4 chars of the value in the column 'released'
```

```
#Order Data by gross revenue

df.sort_values(by=['gross'], inplace=False , ascending = False)
```

```
#Now we can drop all the duplicates

#df['company'].drop_duplicates().sort_values(ascending=False) #Shows all companies 
```

```
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```
# Regression Plot Budget vs Gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"BLUE"})


```

```

df.corr(method='pearson') #There are different ways to define a corr. Pearson is a default. Kendall and Spearman will brung slightly different results
```

```
# High corr between budget the gross

corr_matrix = df.corr(method='pearson') 

sns.heatmap(corr_matrix, annot=True) #Beautifull way to present corr matrix

plt.title('Corr Matrix')

plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show
```

```
#Company is not a numeric value so we can not define corr. Lets give each company an index

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```
corr_matrix = df_numerized.corr(method='pearson') 

sns.heatmap(corr_matrix, annot=True) #Beautifull way to present corr matrix

plt.title('Corr Matrix')

plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show
```

```
df_numerized.corr()
```

```
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
